```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from packaging import markers

from WBUtiles.WBClean import clean_wb_wide
from WBUtiles.WBSummary import summary_wb, plot_missing_heatmap
from WBUtiles.WBLatex import make_lookup_latex, make_summary_latex

sns.set_theme(
    style="whitegrid",     # clean academic look
    context="paper",       # scales text for papers
    font_scale=1.1
)
```

```python
print(sns.color_palette().as_hex())
```

```python
zambia_hc = pd.read_csv('../../data/Zambia_HumanCapital/human_capital.csv')
lic_hc = pd.read_csv('../../data/LIC_HumanCapital/human_capital.csv')
```

```python
zambia_clean, zambia_lookup = clean_wb_wide(zambia_hc, drop_top_rows=0)
lic_clean, lic_lookup = clean_wb_wide(lic_hc, drop_top_rows=0)
```

```python
# plot missing data
availability_zambia, fig_zambia, ax_zambia = plot_missing_heatmap(
    zambia_clean,
    lookup=zambia_lookup
)
availability_lic, fig_lic, ax_lic = plot_missing_heatmap(
    lic_clean,
    lookup=lic_lookup
)
```

# Education

```python
# gross secondary education
df_zambia = zambia_clean.dropna(subset=['SE.SEC.ENRR'])
df_lic = lic_clean.dropna(subset=['SE.SEC.ENRR'])
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(df_zambia['Year'], df_zambia['SE.SEC.ENRR'], marker = 'o', label='Zambia')
ax.plot(df_lic['Year'], df_lic['SE.SEC.ENRR'], label='LIC')
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('% gross population', fontsize=12)

# set range
# ax.set_ylim(700, 4000)
ax.set_xlim(1958, 2026)
ax.grid(False)
ax.legend(fontsize=12)

fig.savefig(
    "../figures/LIC/human_capital/gross_second_edu.pdf",
    format="pdf",
    bbox_inches="tight"
)
```

```python
# gross secondary education gender
df_zambia = zambia_clean.dropna(subset=['SE.SEC.ENRR.MA', 'SE.SEC.ENRR.FE'])
df_lic = lic_clean.dropna(subset=['SE.SEC.ENRR.MA', 'SE.SEC.ENRR.FE'])

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(df_zambia['Year'], df_zambia['SE.SEC.ENRR.MA'],
        marker = 'o', label='Zambia_male', color='#4c72b0', linestyle = '--')
ax.plot(df_zambia['Year'], df_zambia['SE.SEC.ENRR.FE'],
           marker = 'o', label='Zambia_female', color = '#dd8452', linestyle = '--')

ax.plot(df_lic['Year'], df_lic['SE.SEC.ENRR.MA'], label='LIC_male', color='#4c72b0')
ax.plot(df_lic['Year'], df_lic['SE.SEC.ENRR.FE'], label='LIC_female', color = '#dd8452')

ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('% gender gross population', fontsize=12)
ax.legend(fontsize=12)

# set range
# ax.set_ylim(700, 4000)
ax.set_xlim(1958, 2026)
ax.grid(False)

fig.savefig(
    "../figures/LIC/human_capital/gross_second_edu_gender.pdf",
    format="pdf",
    bbox_inches="tight"
)
```

```python
# Life expectancy Gender
df_lic = lic_clean.dropna(subset=['SP.DYN.LE00.MA.IN', 'SP.DYN.LE00.FE.IN'])
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(df_lic['Year'], df_lic['SP.DYN.LE00.MA.IN'], label='Male')
ax.plot(df_lic['Year'], df_lic['SP.DYN.LE00.FE.IN'], label='Female')

ax.fill_between(
    df_lic['Year'],
    df_lic['SP.DYN.LE00.FE.IN'],
    df_lic['SP.DYN.LE00.MA.IN'],
    alpha=0.2)

ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Years', fontsize=12)
ax.set_xlim(1958, 2026)
ax.grid(False)
ax.legend(fontsize=12)

fig.savefig(
    "../figures/LIC/human_capital/life_expectancy_gender.pdf",
    format="pdf",
    bbox_inches="tight"
)
```
